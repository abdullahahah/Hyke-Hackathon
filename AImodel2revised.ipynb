```python
import pandas as pd
from pycaret.regression import *
from pycaret.regression import compare_models

# Load and prepare your data
data = pd.read_csv('supermarket3.csv')

# Setup
regression_setup = setup(data=data, target='Sales', train_size=0.8)


# Compare and evaluate different linear regression models
best_model = compare_models(fold=5, sort='MAE')

# Train the linear regression model on the complete dataset
final_model = create_model(best_model)

# Make predictions on new data
new_data = pd.read_csv('new_data2.csv')  # Assuming you have a new dataset
new_data.head()
predictions = predict_model(final_model, data=new_data)

# The predictions
print(predictions)

evaluate_model(final_model)
```
